# 3. The Fourier transform

While Fourier series are limited to periodic data, the goal of the Fourier transform is generalization to non-periodic data. The idea is to achieve this by extending the period $T_p$ to infinity. In what follows, we will look at the details.

The computation of the Fourier series ({ref}`sec:fourierove_vrste`) is based on the period $T_p$; the integral for computing the coefficients can be defined from $-T_p/2$ to $+T_p/2$:

$$
c_n=\frac{1}{T_p}\,\int_{-T_p/2}^{+T_p/2}x(t)\,\mathrm{e}^{-\textrm{i}\,\omega_{p}\,n\,t}\,\textrm{d} t,
$$

where $\omega_p=2\pi/T_p$ is the fundamental angular frequency. When $T_p$ becomes large and goes to infinity $T_p\rightarrow \infty$, the fundamental angular frequency goes to zero: $\omega_p\rightarrow 0$. To avoid problems with division by $\infty$, we multiply the above expression by $T_p$, use $\omega_n=n\,\omega_p$ and continue:

$$
\lim_{T_p\rightarrow \infty} c_n\,T_p=\underbrace{\int_{-T_p/2}^{+T_p/2}x(t)\,\mathrm{e}^{-\textrm{i}\,\omega_{n}\,t}\,\textrm{d} t}_{X(\omega_n)}.
$$

Assuming that the limit exists, we write the **Fourier transform** (also the **Fourier integral**):

$$
X(\omega)= \mathcal{F}\{x(t)\} = \int_{-\infty}^{+\infty}x(t)\,\mathrm{e}^{-\textrm{i}\,\omega\,t}\,\textrm{d} t,
$$
where instead of $\omega_{n}$ we wrote $\omega$, since the frequency is now continuously defined. $X(\omega)$ represents the amplitude frequency density, since $c_n\,T_p=c_n/\Delta f$, where $\Delta f$ is the frequency bandwidth. We denote the Fourier transform by $\mathcal{F}\{\}$.

Let us now also look at the **reconstructed continuous function**, when $T_p\rightarrow \infty$:

$$
x(t)=\lim_{T_p\rightarrow\infty}\sum_{n=-\infty}^{+\infty}
X(\omega_n)\underbrace{\frac{1}{T_p}}_{\Delta f}\,\mathrm{e}^{\mathrm{i}\,\overbrace{2\pi\,n/T_p}^{\omega_n}\,t}
=\lim_{T_p\rightarrow\infty}\sum_{n=-\infty}^{+\infty}
X(\omega_n)\,\mathrm{e}^{\mathrm{i}\,\omega_n\,t}\Delta f.
$$

The given expression is continuous and we call it the **inverse Fourier transform**. We denote it by $\mathcal{F}^{-1}\{\}$:

$$
x(t)=\mathcal{F}^{-1}\{X(\omega)\}=\frac{1}{2\pi}\int_{-\infty}^{+\infty}
X(\omega)\,\mathrm{e}^{\mathrm{i}\,\omega\,t}\textrm{d}\omega.
$$

In the expression above we used $\textrm{d}f=\textrm{d}\omega/(2\pi)$.

:::{note}

Here we repeat once more the Fourier transform and its inverse, written in **angular frequency**:

$$
X(\omega)=\int_{-\infty}^{+\infty} x(t)\,\mathrm{e}^{-\,\textrm{i}\,\omega\,t}\,\textrm{d} t,
\qquad
x(t)=\frac{1}{2\pi}\int_{-\infty}^{+\infty} X(\omega)\,\mathrm{e}^{\,\textrm{i}\,\omega\,t}\,\textrm{d} \omega.
$$

:::

:::{note}

In numerical methods, however, the frequency is more often defined in **ordinary frequency**; then the Fourier transform and its inverse are defined as:

$$
X(f)=\int_{-\infty}^{+\infty} x(t)\,\mathrm{e}^{-\textrm{i}\,2\pi\,f\,t}\,\textrm{d} t,
\qquad
x(t)=\int_{-\infty}^{+\infty} X(f)\,\mathrm{e}^{\textrm{i}\,2\pi\,f\,t}\,\textrm{d} f.
$$

:::

### The Dirac delta function

First let us look at the Fourier transform of the **Dirac delta function** $\delta(t)$:

$$
\mathcal{F}\{\delta(t)\} = \int_{-\infty}^{+\infty} \delta(t)\,\mathrm{e}^{-\,\textrm{i}\,\omega\,t}\,\textrm{d} t=\mathrm{e}^{-\,\textrm{i}\,\omega\,0}=1.
$$

On the basis of the derivation above, we conclude that the Dirac delta function is infinitely short in time and infinitely wide in the frequency domain (it occupies the entire spectrum, i.e. all frequencies). The reverse would also hold: if we defined the delta function in the frequency domain, it would be infinitely narrow, and infinitely wide in time. We will return to this concept later.

### Symmetric exponentially decaying function

Let us now look at the **symmetric exponentially decaying function** $x(t)$:

$$
x(t)=\mathrm{e}^{-\eta\,|t|},\qquad\textrm{where:}\qquad\eta>0.
$$

Fourier transform:

$$
\mathcal{F}\{x(t)\} = \int_{-\infty}^{+\infty} \mathrm{e}^{-\eta\,|t|}\,\mathrm{e}^{-\,\textrm{i}\,2\pi\,f\,t}\,\textrm{d} t=\frac{2\,\eta}{\eta^{2} + 4 \pi^{2}\,f^{2}}.
$$

We observe that at the frequency 0 Hz the amplitude is $X(0)=2/\eta$, and at the frequency $f_{1/2}=|\eta/(2\pi)|$ it falls to half of the initial amplitude. The parameter $\eta$ defines the time spread; the larger the value of $\eta$, the faster the function will converge to zero in the time domain and the slower in the frequency domain. Similarly to the delta function, we therefore find that a small time spread results in a large frequency spread and vice versa.

Let us now look at an example in time and frequency:

```python
import sympy as sym
t, f = sym.symbols('t, f', real=True)
η = sym.symbols('\\eta', real=True, positive=True)
data = {η: 0.1}
x = sym.exp(-η*sym.Abs(t))
X = sym.fourier_transform(x, t, f)

sym.plot(x.subs(data), (t,-50,50), xlabel='$t$ [s]', ylabel='$x(t)$', title='Time domain')
p1 = sym.plot(sym.re(X.subs(data)), (f,-.1,.1), xlabel='$f$ [Hz]', label='Re$(X(f))$', ylabel='', title='Frequency domain', show=False)
p2 = sym.plot(sym.im(X.subs(data)), (f,-.1,.1), line_color='C1', label='Im$(X(f))$', show=False)
p1.extend(p2)
p1.legend = True
p1.show()
```

In the example above we observe that the symmetric exponential function results in a Fourier transform that has only a non-zero real component. This holds in general for all **even functions**; for **odd functions** the reverse would hold: the imaginary component would be non-zero (figure below). We already found something similar with the Fourier series of even and odd functions (for more, see {ref}`sec:FV_lih_sod`).

```python
import sympy as sym
t, f = sym.symbols('t, f', real=True)
η = sym.symbols('\\eta', real=True, positive=True)
data = {η: 0.1}
x = sym.sign(t)*sym.exp(-η*sym.Abs(t))
X = sym.fourier_transform(x, t, f)

sym.plot(x.subs(data), (t,-50,50), xlabel='$t$ [s]', ylabel='$x(t)$', title='Time domain')
p1 = sym.plot(sym.re(X.subs(data)), (f,-.1,.1), xlabel='$f$ [Hz]', label='Re$(X(f))$', ylabel='', 
              title='Frequency domain', show=False)
p2 = sym.plot(sym.im(X.subs(data)), (f,-.1,.1), line_color='C1', label='Im$(X(f))$', show=False)
p1.extend(p2)
p1.legend = True
p1.show()
```

It is worth emphasizing, however, that in general real signals $x(t)$ become complex after the Fourier transform; their amplitude is an even function and their phase an odd function (figure below; $\theta(t)$ denotes the unit-step function). Time signals can in general be of complex form, and then in the frequency domain we generally do not observe the evenness/oddness of the amplitude or phase (try changing the code below).

```python
import sympy as sym
t, f, d = sym.symbols('t, f, d', real=True)
η = sym.symbols('\\eta', real=True, positive=True)
data = {η: 0.1, d: 10}
x = sym.Heaviside(t-d)*sym.exp(-η*sym.Abs(t-d))
# x = sym.sign(t-d)*sym.exp(-η*sym.Abs(t-d)) # time shift, odd function, zero mean
x2 = sym.Heaviside(t)*sym.exp(-η*sym.Abs(t))
x3 = sym.I*sym.Heaviside(t-2*d)*sym.exp(-η*sym.Abs(t-2*d))  # complex time series; this will be fast
# x3 = sym.I*sym.sign(t-2*d)*sym.exp(-η*sym.Abs(t-2*d))  # complex time series; this will be fast, odd function, zero mean
# x = x + sym.I*sym.sign(t)*sym.exp(-η*sym.Abs(t)) # complex time series; this will take a while
X = sym.fourier_transform(x, t, f)
X2 = sym.fourier_transform(x2, t, f)
X3 = sym.fourier_transform(x3, t, f)

p1 = sym.plot(x.subs(data), (t,-50,50), xlabel='$t$ [s]', ylabel='$x(t)$', 
              line_color='C0', title='Time domain', show=False, label=f'${sym.latex(x)}$')
p2 = sym.plot(x2.subs(data), (t,-50,50), line_color='C1',  label=f'${sym.latex(x2)}$', show=False)
p3 = sym.plot(sym.im(x3).subs(data), (t,-50,50), line_color='C2', label=f'Im(${sym.latex(x3)}$)', show=False)
p1.extend(p2)
p1.extend(p3)
p1.legend = True
p1.show()
p1 = sym.plot(sym.Abs(X.subs(data)), (f,-.1,.1), line_color='C0', xlabel='$f$ [Hz]', ylabel='$|X(f)|$', 
         title='Frequency domain $-$ amplitude', show=False, label=f'${sym.latex(x)}$')
p2 = sym.plot(sym.Abs(X2.subs(data)), (f,-.05,.05), line_color='C1', label=f'${sym.latex(x2)}$', show=False)
p3 = sym.plot(sym.Abs(X3.subs(data)), (f,-.075,.075), line_color='C2', label=f'${sym.latex(x3)}$', show=False)
p1.extend(p3)
p1.extend(p2)
p1.legend = True
p1.show()
p1 = sym.plot(sym.arg(X.subs(data)), (f,-.1,.1), line_color='C0', xlabel='$f$ [Hz]', ylabel='$\\angle X(f)$',
         title='Frequency domain $-$ phase',  show=False, label=f'${sym.latex(x)}$')
p2 = sym.plot(sym.arg(X2.subs(data)), (f,-.1,.1), line_color='C1', label=f'${sym.latex(x2)}$', show=False)
p3 = sym.plot(sym.arg(X3.subs(data)), (f,-.1,.1), line_color='C2', label=f'${sym.latex(x3)}$', show=False)
p1.extend(p2)
p1.extend(p3)
p1.legend = True
p1.show()
```

### Sine function

Let us assume a sine function:

$$
x(t) = A\,\sin(2\pi\,p\,t),
$$

which we rewrite into:

$$
x(t) = \frac{A}{2\,\textrm{i}}\,
\left(\mathrm{e}^{\textrm{i}\,2\pi\,p\,t}-\mathrm{e}^{-\textrm{i}\,2\pi\,p\,t}
\right)
$$

and then continue with the computation of the Fourier transform:

```python
import sympy as sym
A, t, f = sym.symbols('A, t, f', real=True)
p = sym.symbols('p', real=True, positive=True)
i = sym.I
π = sym.pi

x = A*(sym.exp(i*2*π*p*t)-sym.exp(-i*2*π*p*t))/(2*i)
X = sym.fourier_transform(x, t, f)
X
```

In the above symbolic derivation, $\mathcal{F}_{t}[\cdot]$ represents the Fourier transform. Given the properties of the Dirac delta function (see {ref}`sec:dirac_lastnosti`), we know that $\mathcal{F}_{t}\left[\mathrm{e}^{-2\,\textrm{i}\,\pi\,p\,t}\right]\left(f\right)=\delta(p+f)$, from which it follows:

$$
\mathcal{F}_{t}\left( A\,\sin(2\pi\,p\,t)\right)=
\frac{\mathrm{i} A}{2} \delta(f + p) - \frac{\mathrm{i} A}{2}\delta(f - p),
$$

which is shown in the figure below.

```python
import numpy as np
import matplotlib.pyplot as plt

A = 1
p = 10

fig, ax1 = plt.subplots()
plt.title('$|\\mathcal{F}_{t}\\left( A\\,\\sin(2\\pi\\,p\\,t)\\right)|$')

ax1.set_xlabel('$f$')
ax1.set_ylabel('$|\\mathcal{F}()|$')
ax1.plot([-p, p], [A/2, A/2], 'o')
ax1.vlines(0, -0.5*A, A, 'k', lw=1)
ax1.vlines(-p, 0, A/2, 'k', lw=0.5)
ax1.vlines(p, 0, A/2, 'k', lw=0.5)
ax1.hlines(0, -2*p, 2*p, 'k', lw=1)
ax1.annotate('$-p$',
            xy=(-p, -0.5), xycoords='data',
            ha='center', size='large',
            xytext=(-p, -0.1), textcoords='data'
            )
ax1.annotate('$+p$',
            xy=(p, -0.5), xycoords='data',
            ha='center', size='large',
            xytext=(p, -0.1), textcoords='data'
            )
ax1.annotate('$A/2$',
            xy=(-p, A/2), xycoords='data',
            ha='center', size='large',
            xytext=(-p, A/2+.1), textcoords='data'
            )
ax1.annotate('$A/2$',
            xy=(p, A/2), xycoords='data',
            ha='center', size='large',
            xytext=(p, A/2+.1), textcoords='data'
            )
plt.axis('off')
plt.show()
```

### Rectangular impulse

Let us assume a rectangular function:

$$
x(t) =
\begin{cases}
a;&\quad |t|<b,\\
0&\quad \textrm{otherwise,}
\end{cases}
$$

which in the case of the Fourier transform leads to the expression:

$$
X(f)=\int_{-\infty}^{+\infty} x(t)\,\mathrm{e}^{-\textrm{i}\,2\pi\,f\,t}\,\textrm{d} t=
\int_{-b}^{+b} a\,\mathrm{e}^{-\textrm{i}\,2\pi\,f\,t}\,\textrm{d} t=\frac{a \sin\left(2\pi\,b\,f \right)}{\pi\,f},
$$

which we can rewrite into:

$$
X(f)=2\,a\,b\,\frac{\sin\left(2\pi\,b\,f \right)}{2\pi\,b\,f }.
$$

The above expression clearly shows: as $f$ tends to 0, $X(f)$ tends to $2\,a\,b$; otherwise it oscillates in the frequency domain with frequency $1/b$. We can see some characteristics from the figure below: again we observe that the wider the data are in time, the narrower they are in frequency.

```python
import sympy as sym
a, b, f = sym.symbols('a, b, f', real=True, positive=True)
t = sym.symbols('t', real=True)
i = sym.I
π = sym.pi

x = a*sym.Heaviside(t+b) - a*sym.Heaviside(t-b)

X = sym.simplify(sym.integrate(a*sym.exp(-i*2*π*f*t), (t, -b, b)), trig=True)

data = {a: 1, b:1}
data2 = {a: 1, b:3}

p1 = sym.plot(x.subs(data), (t, -5, +5), line_color='C0', label=f'$b={b.subs(data)}$',show=False, 
              title='Time domain', ylabel='$x(t)$', xlabel='$t$')
p2 = sym.plot(x.subs(data2), (t, -5, +5), line_color='C1', label=f'$b={b.subs(data2)}$',show=False)
p1.extend(p2)
p1.legend = True
p1.show()
p1 = sym.plot(X.subs(data), (f, -5, +5), line_color='C0', label=f'$b={b.subs(data)}$',show=False, 
              title='Frequency domain', ylabel='$X(f)$', xlabel='$f$')
p2 = sym.plot(X.subs(data2), (f, -5, +5), line_color='C1', label=f'$b={b.subs(data2)}$',show=False)
p1.extend(p2)
p1.legend = True
p1.show()
```

### Gaussian impulse

The Gaussian window has a very important place in various theoretical derivations (e.g. probability distribution, windows, wavelet transform), and therefore we will look here at its form in the frequency domain. Let us first define:

$$
x(t) = \mathrm{e}^{-a\,t^2};\qquad a>0.

We derive the Fourier transform (below), which results in:

$$
X(f)=\sqrt{\frac{\pi}{a}}\, \mathrm{e}^{-\pi^{2}\,f^{2}/a}.

```python
import sympy as sym
t, f = sym.symbols('t, f', real=True)
a = sym.symbols('a', real=True, positive=True)

x = sym.exp(-a*t**2)
X = sym.fourier_transform(x, t, f)
X
```

```python
import sympy as sym
t, f = sym.symbols('t, f', real=True)
a = sym.symbols('a', real=True, positive=True)
x = sym.exp(-a*t**2)
X = sym.fourier_transform(x, t, f)

data = {a: 1}

p1 = sym.plot(x.subs(data), (t, -2, +2), line_color='C0', 
              label=f'$a={a.subs(data)}$', show=False, 
              title='Time domain', ylabel='$x(t)$', xlabel='$t$')
p1.legend = True
p1.show()

p1 = sym.plot(sym.Abs(X).subs(data), (f, -2, +2), line_color='C0',
              label=f'$a={a.subs(data)}$'+', Max$=\\sqrt{\\pi/a}=$'+f'{sym.sqrt(sym.pi/a).evalf(subs=data):3.2f}', show=False, 
              title='Frequency domain $-$ amplitude', ylabel='$|X(f)|$', xlabel='$f$')
p1.legend = True
p1.show()
```

## Properties of the Fourier transform

### Linearity

:::{note}

The Fourier transform is linear, which means:

$$
\mathcal{F}\{a\,x(t)+b\,y(t)\} = a\,X(f) + b\,Y(f).
$$
:::

And the derivation:

```python
import sympy as sym
t, f = sym.symbols('t, f', real=True)
a, b = sym.symbols('a, b', real=True, positive=True)
x = sym.Function('x')
y = sym.Function('y')
i = sym.I
π = sym.pi

A = sym.integrate(( a*x(t) + b*y(t) )*sym.exp(-i*2*π*f*t), (t, -sym.oo, +sym.oo))
A
```

We expand the expression; since $a$ and $b$ are constants, we confirm linearity:

```python
sym.expand(A)
```

Below, let us look at a concrete example.

```python
import sympy as sym
t, f, a, b = sym.symbols('t, f, a, b', real=True)
x_1 = sym.exp(-t**2)
x_2 = a * x_1 + b*x_1
data = {a: -2, b:4}

X_1 = sym.fourier_transform(x_1, t, f)
X_2 = sym.fourier_transform(x_2.subs(data), t, f)


p1 = sym.plot(x_1.subs(data), (t, -2, +2), line_color='C0', 
              label=f'$x_1(t)$', show=False, 
              title='Time domain', ylabel='', xlabel='$t$')
p2 = sym.plot(x_2.subs(data), (t, -2, +2), line_color='C1', 
              label=f'$x_2(t)=a\\,x_1(t)+b\\,x_1(t), a={a.subs(data)}, b={b.subs(data)}$', show=False)
p1.extend(p2)
p1.legend = True
p1.show()

p1 = sym.plot(sym.Abs(X_1).subs(data), (f, -2, +2), line_color='C0',
              label=f'$|X_1(f)|$', show=False, 
              title='Frequency domain $-$ amplitude', ylabel='', xlabel='$f$')
p2 = sym.plot(sym.Abs(X_2).subs(data), (f, -2, +2), line_color='C1',
              label='$|X_2(f)|$', show=False)
p1.extend(p2)
p1.legend = True
p1.show()
```

### Time scaling

:::{note}

Time scaling is defined as $a\,t$ and has the following effect on the Fourier transform:

$$
\mathcal{F}\{x(a\,t)\} = \frac{1}{|a|}\,X(f/a).
$$
:::

And the derivation (for positive $a$):

```python
import sympy as sym
t, f, τ = sym.symbols('t, f, tau', real=True)
a = sym.symbols('a', real=True, positive=True)
x = sym.Function('x')
i = sym.I
π = sym.pi
X = sym.integrate(x(a*t)*sym.exp(-i*2*π*f*t), (t, -sym.oo, +sym.oo))

sym.Eq(X, X.transform(a*t, τ))
```

Time scaling becomes important, for example, when we observe the same event with two different clocks.

```python
import sympy as sym
t, f, a = sym.symbols('t, f, a', real=True)
x_1 = sym.exp(-t**2)
x_2 = x_1.subs(t, a*t)
data = {a: 1.1}

X_1 = sym.fourier_transform(x_1, t, f)
X_2 = sym.fourier_transform(x_2.subs(data), t, f)


p1 = sym.plot(x_1.subs(data), (t, -2, +2), line_color='C0', 
              label=f'$t=t$', show=False, 
              title='Time domain', ylabel='$x(t)$', xlabel='$t$')
p2 = sym.plot(x_2.subs(data), (t, -2, +2), line_color='C1', 
              label=f'$t=a\\,t$, $a=${a.subs(data):3.2f}', show=False)
p1.extend(p2)
p1.legend = True
p1.show()

p1 = sym.plot(sym.Abs(X_1).subs(data), (f, -2, +2), line_color='C0',
              label=f'$t=t$', show=False, 
              title='Frequency domain $-$ amplitude', ylabel='$|X(f)|$', xlabel='$f$')
p2 = sym.plot(sym.Abs(X_2).subs(data), (f, -2, +2), line_color='C1',
              label=f'$t=a\\,t$, $a=${a.subs(data):3.2f}', show=False)
p1.extend(p2)
p1.legend = True
p1.show()
```

### Time reversal

Time reversal has the following effect on the Fourier transform:

\mathcal{F}\{x(-t)\} = X(-f)\qquad \left(=X^*(f), \quad\textrm{if } x(t)\textrm{ is a real function}\right).
$$

```python
import sympy as sym
t, f, τ = sym.symbols('t, f, tau', real=True)
a = sym.symbols('a', real=True, positive=True)
x = sym.Function('x')
i = sym.I
π = sym.pi
X = sym.integrate(x(-t)*sym.exp(-i*2*π*f*t), (t, -sym.oo, +sym.oo))

sym.Eq(X, X.transform(-t, τ))
```

### Time shifting

:::{note}

The time shift is defined by $t_0$ and has the following effect on the Fourier transform:

$$
\mathcal{F}\{x(t-t_0)\} = \mathrm{e}^{-\textrm{i}\,2\pi\,f\,t_0}\,X(f).
$$
:::

And the derivation:

```python
import sympy as sym
t, t_0, f, τ = sym.symbols('t, t_0, f, tau', real=True)
a = sym.symbols('a', real=True, positive=True)
x = sym.Function('x')
i = sym.I
π = sym.pi
X = sym.integrate(x(t-t_0)*sym.exp(-i*2*π*f*t), (t, -sym.oo, +sym.oo))

sym.Eq(X, X.transform(t-t_0, τ))
```

We observe that the term $\mathrm{e}^{-\textrm{i}\,2\pi\,f\,t_0}$ can be factored out of the integral.

Time shifting becomes important when we deal with delayed measurements (e.g. if we measure one signal with a delay relative to the others): see the example below.

```python
import sympy as sym
t, t_0, f, a = sym.symbols('t, t_0, f, a', real=True)
i = sym.I
π = sym.pi

x_1 = sym.exp(-t**2)
x_2 = x_1.subs(t, t-t_0)
data = {t_0: 1.1}

X_1 = sym.fourier_transform(x_1, t, f)
X_2 = sym.fourier_transform(x_2.subs(data), t, f)
X_3 = X_2/sym.exp(-i*2*π*f*t_0)


p1 = sym.plot(x_1.subs(data), (t, -2, +2), line_color='C0', 
              label=f'$t=t$', show=False, 
              title='Time domain', ylabel='$x(t)$', xlabel='$t$')
p2 = sym.plot(x_2.subs(data), (t, -2, +2), line_color='C1', 
              label=f'$t=t-t_0$, $t_0=${t_0.subs(data):3.2f}', show=False)
p1.extend(p2)
p1.legend = True
p1.show()

p1 = sym.plot(sym.Abs(X_1).subs(data), (f, -2, +2), line_color='C0',
              label=f'$t=t$', show=False, 
              title='Frequency domain $-$ amplitude', ylabel='$|X(f)|$', xlabel='$f$')
p2 = sym.plot(sym.Abs(X_2).subs(data), (f, -2, +2), line_color='C1',
              label=f'$t=t-t_0$, $t_0=${t_0.subs(data):3.2f}', show=False)
p1.extend(p2)
p1.legend = True
p1.show()

p1 = sym.plot(sym.arg(X_1).subs(data), (f, -2, +2), line_color='C0',
              label=f'$t=t$', show=False, 
              title='Frequency domain $-$ phase', ylabel='$\\angle X(f)$', xlabel='$f$')
p2 = sym.plot(sym.arg(X_2).subs(data), (f, -2, +2), line_color='C1',
              label=f'$t=t-t_0$, $t_0=${t_0.subs(data):3.2f}', show=False)
p1.extend(p2)
p1.legend = True
p1.show()

p1 = sym.plot(sym.arg(X_3).subs(data), (f, -2, +2), line_color='C2',
              label=f'$t=t$', show=True, ylim=[-3.2, 3.2],
              title='Frequency domain $-$ phase $-$ correction: $\\mathcal{F}\\{x(t-t_0)\\} / \\mathrm{e}^{-i\\,2\\pi\\,f\\,t_0}$', ylabel='$\\angle X(f)$', xlabel='$f$')
p1.legend = True
```

### Amplitude modulation (also *frequency shifting*)

:::{note}

Amplitude modulation is defined by $\mathrm{e}^{\textrm{i}\,2\pi\,f_0\,t}$ and has the following effect on the Fourier transform:

$$
\mathcal{F}\{x(t)\,\mathrm{e}^{\textrm{i}\,2\pi\,f_0\,t}\} = X(f-f_0).
$$

We also encounter amplitude modulation in the form multiplied by a cosine function:

$$
\mathcal{F}\{x(t)\,\cos(2\pi\,f_0\,t)\} = \frac{1}{2}\big(X(f-f_0)+X(f+f_0)\big).
$$

:::

And the derivation:

```python
import sympy as sym
t, f, τ = sym.symbols('t, f, tau', real=True)
f_0 = sym.symbols('f_0', real=True, positive=True)
x = sym.Function('x')
i = sym.I
π = sym.pi
X = sym.integrate(x(t)*sym.exp(i*2*π*f_0*t)*sym.exp(-i*2*π*f*t), (t, -sym.oo, +sym.oo))

sym.Eq(X, sym.simplify(X))
```

Amplitude modulation is relatively common in practice. We observe it, for example, when we measure the vibration of a plate with a laser and the laser's base shakes; in the figures below, the displacement of the measured structure is denoted by $x(t)$, while the laser's base vibrates with $\cos(2\pi\,f_0\,t)$.

```python
import sympy as sym
t, f = sym.symbols('t, f', real=True)
f_0 = sym.symbols('f_0', real=True, positive=True)
i = sym.I
π = sym.pi

x_1 = sym.exp(-t**2)
x_2 = x_1*sym.cos(2*π*f_0*t)
data = {f_0: 1}

X_1 = sym.fourier_transform(x_1, t, f)
X_2 = sym.fourier_transform(x_2.subs(data), t, f)

X_1 = sym.fourier_transform(x_1, t, f)
X_2 = sym.fourier_transform(x_2.subs(data), t, f)


p1 = sym.plot(x_1.subs(data), (t, -2, +2), line_color='C0', 
              label=f'$x(t)$', show=False, 
              title='Time domain', ylabel='$x(t)$', xlabel='$t$')
p2 = sym.plot(x_2.subs(data), (t, -2, +2), line_color='C1', 
              label=f'$x(t)\\,\\cos(2\\pi\\,f_0\\,t)$', show=False)
p1.extend(p2)
p1.legend = True
p1.show()


p1 = sym.plot(sym.Abs(X_1).subs(data), (f, -2, +2), line_color='C0',
              label='$\\mathcal{F}\\{x(t)\\}$', show=False, 
              title='Frequency domain $-$ amplitude', ylabel='$|X(f)|$', xlabel='$f$')
p2 = sym.plot(sym.Abs(X_2).subs(data), (f, -2, +2), line_color='C1',
              label='$\\mathcal{F}\\{x(t)\\,\\cos(2\\pi\\,f_0\\,t)\\}$', show=False)
p1.extend(p2)
p1.legend = True
p1.show()
```

### Differentiation

:::{note}

If $\dot{x}(t)$ is the time derivative of $x(t)$, their relation in the frequency domain is:

$$
\mathcal{F}\{\dot x(t)\} = \textrm{i}\,2\pi\,f\,X(f),
$$

where it is assumed that $x(t)\rightarrow 0$ as $t\rightarrow \pm\infty$.

:::

And the derivation:

```python
import sympy as sym
t, f = sym.symbols('t, f', real=True)
x = sym.Function('x')
i = sym.I
π = sym.pi

X = sym.integrate(sym.diff(x(t),t)*sym.exp(-i*2*π*f*t), (t, -sym.oo, +sym.oo))
X
```

We must use integration by parts $u = \mathrm{e}^{-\textrm{i}\,2\pi\,f\,t}$, $\textrm{d}v= \dot{x}(t)\,\textrm{d}t$; we prepare $\textrm{d}u$ and $v$:

```python
u = sym.exp(-i*2*π*f*t)
dv = sym.diff(x(t),t)

du = sym.diff(u, t)
v = sym.integrate(dv, t)
```

It follows that we must (the first part of integration by parts):

```python
u*v
```

evaluate at the limits $+\infty$ and $-\infty$; according to the definition above, in both cases (because of the earlier assumption) $x(t=\pm \infty)=0$.

Therefore only the second part of integration by parts remains:

```python
sym.integrate(v*du, (t, -sym.oo, +sym.oo))
```

and with this we have completed the proof.

### Integration

:::{note}

If $\int_{-\infty}^t x(\tau)\,\textrm{d}\tau$ is the integral of the function $x(t)$, their relation in the frequency domain is:

$$
\mathcal{F}\left\{\int_{-\infty}^t x(\tau)\,\textrm{d}\tau \right\} = \frac{X(f)}{\textrm{i}\,2\pi\,f}.
$$

:::

The derivation for integration follows from the one for differentiation:

$$
\mathcal{F}\{\dot{x}(t)\} = \textrm{i}\,2\pi\,f\,\mathcal{F}\{x(t)\},
$$

where we make the substitution of the function: $u(t)=\dot{x}(t)$ and $\int_{-\infty}^t u(\tau)\,\textrm{d}\tau=x(t)$.

Differentiation and integration in the frequency domain are very important properties of the Fourier transform, since they allow us a simple transition between differentiated and integrated data.

Here it is worth emphasizing that time differentiation is an uncertain numerical operation, since it is based on a small difference of numbers (and the difference is then divided by a small number, which further increases the uncertainty). Differentiation in the frequency domain is not exposed to such uncertainty.

Furthermore, we emphasize that time integration is an unstable numerical operation, since the error of an individual step accumulates and therefore grows with time. Integration in the frequency domain is not exposed to such instability; it is true, however, that at the frequency 0 Hz we have an undefined result (division by 0).

```python
import sympy as sym
t, f = sym.symbols('t, f', real=True)
i = sym.I
π = sym.pi

x_1 = sym.exp(-t**2)
x_2 = sym.diff(x_1, t)
x_3 = sym.integrate(x_2, t)

X_1 = sym.fourier_transform(x_1, t, f)
X_2 = sym.fourier_transform(x_2, t, f)
X_3 = sym.fourier_transform(x_3, t, f)

p1 = sym.plot(x_1, (t, -2, +2), line_color='C0', 
              label=f'$x_1(t)={sym.latex(x_1)}$', show=False,
              title='Time domain', ylabel='$x(t)$', xlabel='$t$')
p2 = sym.plot(x_2, (t, -2, +2), line_color='C1', 
              label=f'$x_2(t)=\\dot x_1(t)={sym.latex(x_2)}$', show=False)
p3 = sym.plot(x_3, (t, -1, +1), line_color='C2', 
              label=f'$x_3(t)=\\int x_2(t)dt={sym.latex(x_3)}$', show=False)
p1.extend(p2)
p1.extend(p3)
p1.legend = True
p1.show()

p1 = sym.plot(sym.Abs(X_1), (f, -2, +2), line_color='C0',
              label=f'$|X_1(f)|$', show=False, 
              title='Frequency domain $-$ amplitude', ylabel='$|X(f)|$', xlabel='$f$')
p2 = sym.plot(sym.Abs(X_2), (f, -2, +2), line_color='C1',
              label=f'$|X_2(f)|$', show=False)
p3 = sym.plot(sym.Abs(X_3), (f, -.25, +.25), line_color='C2',
              label=f'$|X_3(f)|$', show=False)
p1.extend(p2)
p1.extend(p3)
p1.legend = True
p1.show()

p1 = sym.plot(sym.arg(X_1), (f, -2, +2), line_color='C0',
              label=f'$\\angle X_1(f)$', show=False, 
              title='Frequency domain $-$ phase', ylabel='$\\angle X(f)$', xlabel='$f$')
p2 = sym.plot(sym.arg(X_2), (f, -2, +2), line_color='C1',
              label=f'$\\angle X_2(f)$', show=False)
p3 = sym.plot(sym.arg(X_3), (f, -1, +1), line_color='C2',
              label=f'$\\angle X_3(f)$', show=False)
p1.extend(p2)
p1.extend(p3)
p1.legend = True
p1.show()

p1 = sym.plot(sym.Abs(X_1), (f, -2, +2), line_color='C0',
              label=f'$|X_1(f)|$', show=False, 
              title='Frequency domain $-$ amplitude', ylabel='$|X(f)|$', xlabel='$f$')
p2 = sym.plot(sym.Abs(X_2/(2*i*π*f)), (f, -2, +2), line_color='C1',
              label=f'$|X_2(f)|$ with integration in the frequency domain', show=False)
p1.extend(p2)
p1.legend = True
p1.show()

p1 = sym.plot(sym.arg(X_1), (f, -2, +2), line_color='C0',
              label=f'$\\angle X_1(f)$', show=False, 
              title='Frequency domain $-$ phase', ylabel='$\\angle X(f)$', xlabel='$f$')
p2 = sym.plot(sym.arg(X_2/(2*i*π*f)), (f, -2, +2), line_color='C1',
              label=f'$\\angle X_2(f)$ with integration in the frequency domain', show=False)
p1.extend(p2)
p1.legend = True
p1.show()
```

### Convolution of functions

:::{note}

The Fourier transform of the convolution of two functions is equal to the product of the Fourier transforms:

$$
\mathcal{F}\left\{x(t)*y(t)\right\} = X(f)\,Y(f).
$$

:::

The symbol $*$ denotes convolution:

$$
\int_{-\infty}^{+\infty}\big(x(t)*y(t)\big)\,\mathrm{e}^{-\textrm{i}\,2\pi\,f\,t}\textrm{d}t=\int_{-\infty}^{+\infty}\int_{-\infty}^{+\infty} x(\tau)\,y(t-\tau)\,\textrm{d}\tau\,\,\mathrm{e}^{-\textrm{i}\,2\pi\,f\,t}\,\textrm{d}t.
$$

We derive the above relation:

```python
import sympy as sym
t, f, τ, v = sym.symbols('t, f, tau, v', real=True)
x = sym.Function('x')
y = sym.Function('y')
i = sym.I
π = sym.pi

X = sym.integrate(
        sym.integrate(x(τ)*y(t-τ)*sym.exp(-i*2*π*f*t), (τ, -sym.oo, +sym.oo)),
        (t, -sym.oo, +sym.oo))
X
```

We also introduce a new variable:

```python
X = X.transform(t-τ, v)
X
```

Above, it is obvious that we can separate the integral over $\tau$ and $v$, and the separated integral represents the Fourier transforms $X(f)$ and $Y(f)$.

The Fourier transform of a convolution is perhaps the most useful property of the Fourier transform, since multiplication in the frequency domain significantly speeds up the computation of the convolution of two functions. We will make good use of this property in the discrete Fourier transform, while the simple example below shows the equivalence of the result in time versus the one in frequency.

```python
import sympy as sym
t, f, τ = sym.symbols('t, f, τ', real=True)
i = sym.I
π = sym.pi

x_1 = sym.exp(-t**2)
x_2 = sym.integrate(x_1.subs(t,τ)*x_1.subs(t,t-τ), (τ, -sym.oo, +sym.oo)) # convolution in time

X_1 = sym.fourier_transform(x_1, t, f)
X_2 = sym.fourier_transform(x_2, t, f)

X_3 = X_1*X_1 # product in the frequency domain
x_3 = sym.inverse_fourier_transform(X_2, f, t)

p1 = sym.plot(x_1, (t, -2, +2), line_color='C0', 
              label=f'$x_1(t)={sym.latex(x_1)}$', show=False,
              title='Time domain', ylabel='$x(t)$', xlabel='$t$')
p2 = sym.plot(x_2, (t, -2, +2), line_color='C1', 
              label=f'$x_2(t)=x_1(t)*x_1(t)$', show=False)
p3 = sym.plot(x_3, (t, -1., +1.), line_color='C2', 
              label='$x_3(t)=\\mathcal{F}^{-1}(\\,X_1(f)\\,X_1(f)\\,)$', show=False)
p1.extend(p2)
p1.extend(p3)
p1.legend = True
p1.show()

p1 = sym.plot(sym.Abs(X_1), (f, -2, +2), line_color='C0',
              label=f'$|X_1(f)|$', show=False, 
              title='Frequency domain $-$ amplitude', ylabel='$|X(f)|$', xlabel='$f$')
p2 = sym.plot(sym.Abs(X_2), (f, -2, +2), line_color='C1',
              label=f'$|X_2(f)|$', show=False)
p3 = sym.plot(sym.Abs(X_1*X_1), (f, -.25, +.25), line_color='C2',
              label=f'$|X_1(f)\\,X_1(f)|$', show=False)
p1.extend(p2)
p1.extend(p3)
p1.legend = True
p1.show()
```

### Product of functions

:::{note}

The Fourier transform of the product of two functions represents, in the frequency domain, the convolution of the Fourier transforms:

$$
\mathcal{F}\left\{x(t)\,y(t)\right\} = X(f)*Y(f).
$$

:::

We derive the above relation:

$$
\int_{-\infty}^{+\infty} \underbrace{x(t)}_{\int_{-\infty}^{+\infty} X(f_1)\,\mathrm{e}^{\mathrm{i}\,2\pi\,t\,f_1}\, \mathrm{d}f_1}\, \overbrace{y(t)}^{\int_{-\infty}^{+\infty} Y(f_2)\,\mathrm{e}^{\mathrm{i}\,2\pi\,t\,f_2}\, \mathrm{d}f_2}\,\mathrm{e}^{-\mathrm{i}\,2\pi\,f\,t}\, \mathrm{d}t=
\int_{-\infty}^{+\infty} X(f_1)\,\int_{-\infty}^{+\infty} Y(f_2)\, \int_{-\infty}^{+\infty} \mathrm{e}^{-\mathrm{i}\,2\pi\,(f-f_1-f_2)\,t}\, \mathrm{d}t\,\mathrm{d}f_2\,\mathrm{d}f_1.
$$

Now we take into account $\int_{-\infty}^{+\infty}\mathrm{e}^{\pm\textrm{i}\,2\pi\,a\,t}\,\textrm{d}t= \delta(a)$ (see {ref}`sec:dirac_lastnosti`) and prove the above derivation:

$$
\int_{-\infty}^{+\infty} X(f_1)\,\underbrace{\int_{-\infty}^{+\infty} Y(f_2)\, \delta(f-f_1-f_2)\,\mathrm{d}f_2}_{Y(f-f_1)}\,\mathrm{d}f_1=\int_{-\infty}^{+\infty} X(f_1)\,Y(f-f_1)\,\mathrm{d}f_1= X(f)*Y(f).
$$

Below, let us look at a concrete example.

```python
import sympy as sym
t, f, f_1 = sym.symbols('t, f, f_1', real=True)
i = sym.I
π = sym.pi

x_1 = sym.exp(-t**2)
x_2 = x_1*x_1
X_1 = sym.fourier_transform(x_1, t, f)
X_2 = sym.fourier_transform(x_2, t, f) # product in time

X_3 = sym.integrate(X_1.subs(f, f_1)*X_1.subs(f, f-f_1), (f_1, -sym.oo, +sym.oo)) # convolution in the frequency domain
x_3 = sym.inverse_fourier_transform(X_3, f, t)

p1 = sym.plot(x_1, (t, -2, +2), line_color='C0', 
              label=f'$x_1(t)={sym.latex(x_1)}$', show=False,
              title='Time domain', ylabel='$x(t)$', xlabel='$t$')
p2 = sym.plot(x_2, (t, -2, +2), line_color='C1', 
              label=f'$x_2(t)=x_1(t)\\,x_1(t)$', show=False)
p3 = sym.plot(x_3, (t, -1., +1.), line_color='C2', 
              label='$x_3(t)=\\mathcal{F}^{-1}(\\,X_1(f)*X_1(f)\\,)$', show=False)
p1.extend(p2)
p1.extend(p3)
p1.legend = True
p1.show()

p1 = sym.plot(sym.Abs(X_2), (f, -2, +2), line_color='C0',
              label=f'$|X_2(f)|$', show=False, 
              title='Frequency domain $-$ amplitude', ylabel='$|X(f)|$', xlabel='$f$')
p2 = sym.plot(sym.Abs(X_3), (f, -.5, +.5), line_color='C1',
              label=f'$|X_3(f)|$', show=False)
p1.extend(p2)
p1.legend = True
p1.show()
```

## The use of windows

Let us assume that we want to compute the Fourier transform of a deterministic signal $x(t)$, which is in principle defined from $-\infty$ to $+\infty$, but which we observe only on the interval from $-T/2$ to $+T/2$. We obtain the following data:

$$
x_T(t)=x(t)\,w(t),
$$

where:

$$
w(t)=
\begin{cases}
1\quad&\textrm{for:}\quad |t|\le T/2,\\
0\quad&\textrm{otherwise}.
\end{cases}
$$

If we obtain the Fourier transform of $x_T(t)$, then:

$$
X_T(f)=\mathcal{F}\{x(t)\,w(t)\}= X(f)*W(f)=\int_{-\infty}^{+\infty} X(g)\,W(f-g)\,\textrm{d}g.
$$

The message, then, is: $X(f)$ represents the frequency transform of the function $x(t)$, which we can observe only in a certain time-limited window $w(t)$, and this window causes the Fourier transform to be distorted. We have already looked at the rectangular impulse $w(t)$ above ({ref}`sec:pravokotni_impulz`) and found that its frequency transform is defined as:

$$
W(f)=T\,\frac{\sin\left(\pi\,T\,f \right)}{\pi\,T\,f },
$$

where we used a notation that reveals the limit: as $T\to 0$, $W(0)=T$.

```python
import numpy as np
import matplotlib.pyplot as plt

A = 1
T = 1.
N = 1000
t = np.linspace(-T/2, T/2, N, endpoint=False)
dt = t[1] - t[0]
# rectangular
wp = np.ones_like(t)
Tp = 0.5*T/2 # theoretically we should take 1*T/2 here
wp[np.abs(t)>Tp] = 0
fT_teor = np.arange(-10,10,0.01)
Wp_teor = A*T / (np.pi*fT_teor) * np.sin(np.pi*fT_teor)

# Bartlett
wBartlett = np.ones_like(t)
wBartlett = A*(wBartlett-np.abs(t)/(T/2))
WBartlett_teor = 0.5*A*T * np.power(np.sin(0.5*np.pi*fT_teor) / (0.5*np.pi*fT_teor),2)

# Hann
wHann = A*np.power(np.cos(np.pi*t/T), 2)
WHann_teor = 0.5*A*T * np.sin(np.pi*fT_teor) / (np.pi*fT_teor*(1-fT_teor**2))

# Hamming
wHamming = 0.54+0.46*np.cos(2*np.pi*t/T)
WHamming_teor = (0.54*np.pi**2-0.08*(np.pi*fT_teor)**2)*np.sin(np.pi*fT_teor)/(np.pi*fT_teor*(np.pi**2-(np.pi*fT_teor)**2))

plt.plot(t,wp,'C0', label='Rectangular $-$ we simulate T/4!')
plt.plot(t,wBartlett,'C1', label='Bartlett')
plt.plot(t,wHann,'C2', label='Hann')
plt.plot(t,wHamming,'C3', label='Hamming')
plt.title('Different windows in the time domain')
plt.xlabel('$t$ [$T$]')
plt.ylabel('Amplitude')
plt.legend(loc=(1.01, 0));
plt.show()

draw = plt.plot
# draw = plt.semilogy # with this option ylim needs to be fixed
plt.vlines(np.arange(-5,6), 0,1, 'k', lw=1, alpha=0.5)
draw(fT_teor, Wp_teor, 'C0', label='Rectangular')
draw(fT_teor, WBartlett_teor, 'C1', label='Bartlett')
draw(fT_teor, WHann_teor, 'C2', label='Hann')
draw(fT_teor, WHamming_teor, 'C3', label='Hamming')
plt.xlim(-5,5)
plt.title('Different windows in the frequency domain')
plt.xlabel('$f$ [1/T]')
plt.ylabel('Amplitude')
plt.legend(loc=(1.01, 0));
```

For a sinusoid of frequency `fr` with amplitude 1, we know that in the frequency domain it has amplitude 1/2 at the frequencies `-fr` and `fr`; on the other hand, above we defined the frequency function of the rectangular window. How the rectangular window distorts the sinusoid can be observed by convolving their frequency transforms (example below).

```python
import sympy as sym
t, f, τ, fr, T = sym.symbols('t, f, τ, fr, T', real=True)
i = sym.I
π = sym.pi

data = {fr: 2.5, T: 0.5}

x = sym.sin(2*π*fr*t)
w = sym.Heaviside(t+T/2) - sym.Heaviside(t-T/2)

xw = x*w
X = 1/2 * sym.DiracDelta(f+fr) + 1/2 * sym.DiracDelta(f-fr)
W = T / (π*f) * sym.sin(π*f)


XW = sym.integrate(X.subs(f,τ)*W.subs(f,f-τ), (τ, -sym.oo, +sym.oo)) # convolution in the frequency domain

p1 = sym.plot(x.subs(data), (t, -2, +2), line_color='C0', 
              label=f'$x(t)=sin(2\\pi\\,fr\\,t)$', show=False,
              title='Time domain', ylabel='', xlabel='$t$ [s]')
p2 = sym.plot(w.subs(data), (t, -2, +2), line_color='C1', 
              label=f'$w(t)$, rectangular window $T={T.subs(data):3.1f}$', show=False)
p1.extend(p2)
p1.legend = True
p1.show()

p1 = sym.plot(0.5*(sym.Heaviside(f+(fr+0.05))-sym.Heaviside(f+(fr-0.05))+
                   sym.Heaviside(f-(fr-0.05))-sym.Heaviside(f-(fr+0.05))).subs(data), 
              (f, -5, +5), line_color='C0', 
              label=f'$X(f)$', show=False,
              title='Frequency domain', ylabel='', xlabel='$f$ [Hz]')
p2 = sym.plot(W.subs(data), (f, -5, +5), line_color='C1', 
              label=f'$W(f)$', show=False)
p3 = sym.plot(XW.subs(data), (f, -5, +5), line_color='C2', 
              label=f'$X(f)\\,W(f)$', show=False)
p1.extend(p2)
p1.extend(p3)
p1.legend = True
p1.show()
```
